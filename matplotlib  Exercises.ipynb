```python
import numpy as np
import pandas  as pd
import  matplotlib.pyplot as plt
%matplotlib inline
x=np.array([0,6])
y=np.array([10,250])
plt.plot(x,y,color="r")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Example')
plt.show()
```

```python
x1=np.linspace(-np.pi,np.pi,256)
x2=np.linspace(-np.pi,np.pi,256)
y1=np.cos(x1)
y2=np.sin(x2)
plt.figure(figsize=(10,6),dpi=80)
plt.plot(x1,y1,'r*--')
plt.plot(x2,y2,'b--')
plt.show()
```

create multiplots on same canvas

```python
x=np.linspace(0,5,11)
y=x**2
z=x**3
print(x)
plt.subplot(1,3,1)#(nrows,ncolumns,plot_number)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)#plot number two
plt.plot(y,x,'g*-')
plt.subplot(1,3,3)
plt.plot(x,z,'y*-')
plt.show()
```

```python
#create figure(empty canvas)
fig=plt.figure()
#add set of axes to figure
axes=fig.add_axes([0.1,0.1,0.8,0.8])#left,bottom,width,height(range 0 to 1)
#plot on that set axes
axes.plot(x,y,'b--')
axes.set_xlabel("set x label")
axes.set_ylabel("set y label")
axes.set_title('set title')
plt.show()
```

You can create a plot within a plot

```python
#create figure(empty canvas)
fig=plt.figure()
#add set of axes to figure
axes1=fig.add_axes([0.1,0.1,0.8,0.8])#main axes
axes2=fig.add_axes([0.2,0.5,0.4,0.3])#inset axes
#plot on that set axes
#Larger figure Axes1
axes1.plot(x,y,'b--')
axes1.set_xlabel("set x label")
axes1.set_ylabel("set y label")
axes1.set_title('set title')
plt.show()
#Insert Figure Axes 2
axes2.plot(y,x,'r')
axes2.set_xlabel("set x1 label")
axes2.set_ylabel("set y1 label")
axes2.set_title('set title1')
plt.show()
```

subplot()-Object will act as a more automatic axis manager

Use similar to plt.figure() except use tuple unpacking to grab fig and axes 



```python
fig,axes=plt.subplots()
axes.plot(x,y,'r')
axes.set_xlabel('x axis')
axes.set_ylabel('y axis')
axes.set_title('title')
plt.show()
```

```python
x=np.linspace(1,5,5)
y=x*2
fig,axes=plt.subplots(1,5)
for i in axes:
    i.plot(x,y,label='xy''r')
    i.set_xlabel('x axis')
    i.set_ylabel('y axis')
    i.set_title('title')

fig #Display the  figure object 
```

**figure size ,aspect ratio and DPI**

You can use the,

figsize and dpi keyword arguments
    
*figsize is a tuple of the width and  height of the  figure in inches
    
*dpi-dot per inch    

*legends

You can use the label='label text' keyword arguments when plots or objects are added to the figure

```python
x1=np.linspace(-np.pi,np.pi,256)
x2=np.linspace(-np.pi,np.pi,256)
y1=np.cos(x1)
y2=np.sin(x2)
plt.figure(figsize=(6,6),dpi=100)
plt.plot(x1,y1,label='x1y1')
plt.plot(x2,y2,label='x2y2')
plt.legend()
plt.show()
```
