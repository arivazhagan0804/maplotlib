```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x=np.arange(0,100)
y=x*2
z=x**2
print(x)
print(y)
print(z)
```

```python
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('title')
plt.show()
```

```python
fig=plt.figure()#create empty canvas
ax1=fig.add_axes([0.1,0.1,0.8,0.8])#left,bottom,width,height
ax2=fig.add_axes([0.2,0.5,0.3,0.3])
ax1.plot(x,y,'r')
ax1.set_xlabel('x axis')
ax1.set_ylabel('y axis')
ax1.set_title('title')
plt.show()
ax2.plot(x,z,'r')
ax2.set_xlabel('x axis')
ax2.set_ylabel('y axis')
ax2.set_title('title')
plt.show()
```

```python
fig=plt.figure(figsize=(12,2),dpi=100)
plt.subplot(1,2,1)
plt.plot(x,y,'b--')
plt.subplot(1,2,2)
plt.plot(x,z,'r-')
plt.show()
```

```python
fig=plt.figure(figsize=(10,2),dpi=200)
plt.subplot(1,2,1)
plt.plot(x,y,'r*-')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('title 1')
plt.subplot(1,2,2)
plt.plot(y,x,'y*-')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('title 2')
plt.show()
```

```python
fig,axes=plt.subplots(1,10,figsize=(4,5),dpi=150)
plt.plot(x,y,'g*')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('title')
plt.show()
```

```python
fig=plt.figure()
ax1=fig.add_axes([0.1,0.1,1,1])
ax2=fig.add_axes([0.2,0.5,0.4,0.5])
ax1.plot(x,y,'b*--')
ax2.plot(y,x,'g*--')
plt.show()
```

```python
from mpl_toolkits import mplot3d
x=np.linspace(0,5,11)
y=np.sin(x)
z=np.cos(x)
fig=plt.figure(figsize=(10,6),dpi=150)
ax=plt.axes(projection='3d')
ax.surface(x,y,z)
plt.show()
```
